```python
import pandas as pd
path=r"E:\project 2\Luxury_Housing_Bangalore.csv"
df=pd.read_csv(path)
df
```

```python
df['Micro_Market']=df['Micro_Market'].str.title().str.strip()
df
```

```python
df.info()
```

```python
df['Ticket_Price_Cr']=df['Ticket_Price_Cr'].astype(str).str.replace(' Cr','', regex=False).str.replace('₹','',regex=False).astype(float)
```

```python
df['Amenity_Score']=df.groupby('Micro_Market')['Amenity_Score'].transform(lambda x: x.fillna(x.mean()))
```

```python
df['Configuration']=df['Configuration'].astype(str).str.lower()
```

```python
df['NRI_Buyer']=df['NRI_Buyer'].map({'yes':True, 'no':False})
```

```python
df['Buyer_Comments']=df['Buyer_Comments'].fillna('no commands')
```

```python
df=df[(df['Ticket_Price_Cr']>0.5) & (df['Ticket_Price_Cr']<60)]
```

```python
df=df[df['Unit_Size_Sqft']>0]
```

```python
df['Price_per_sqft']= (df['Ticket_Price_Cr']*1e7)/df['Unit_Size_Sqft']
```
